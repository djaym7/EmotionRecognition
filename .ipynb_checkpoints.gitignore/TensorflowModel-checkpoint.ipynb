```python
import tensorflow as tf
import numpy as np
import pickle
from sklearn.metrics import confusion_matrix
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def loadData(id):
    with open(f'y_train_{id}','rb') as f:
        y_train=np.array(pickle.load(f))
    with open(f'y_test_{id}','rb') as f:
        y_test=np.array(pickle.load(f))
    with open(f'X_train_{id}','rb') as f:
        X_train = np.array(pickle.load(f))
    with open(f'X_test_{id}','rb') as f:
        X_test = np.array(pickle.load(f))

    

    from sklearn.preprocessing import LabelBinarizer
    encoder = LabelBinarizer()
    y_train = encoder.fit_transform(y_train)
    y_test = encoder.fit_transform(y_test)


    print('X_train:',X_train.shape)
    print('X_test:',X_test.shape)
    print('y_train',y_train.shape) 
    print('y_test',y_test.shape) 

   
    return X_train,y_train,X_test,y_test
```

```python

total_num_people=[]
for i in range(1,92):
    if i <10:
        total_num_people.append('100'+str(i))
    else:
        total_num_people.append('10'+str(i))
epochs = 80
batch_size = 370
results = []
confusionMatrix=[]
for i in total_num_people:
    X_train=[]
    X_test=[]
    y_train=[]
    y_test=[]
    X_train,y_train,X_test,y_test=loadData(i)
    

################################################### MODEL ##################################################################
    from attention import attention
    # Trying attention

    X = tf.placeholder(dtype=tf.float32,name='X',shape=(None,369,20))
    y = tf.placeholder(dtype=tf.int32,name='y',shape=(None,6))


    cell0 = tf.nn.rnn_cell.LSTMCell(num_units=64,activation=tf.nn.tanh,dtype=tf.float32,name='lstm0')
    cell0= tf.nn.rnn_cell.DropoutWrapper(cell0,output_keep_prob=0.5)
    output0,states0 = tf.nn.dynamic_rnn(cell0,X,dtype=tf.float32,swap_memory=True)


    cell1 = tf.nn.rnn_cell.LSTMCell(num_units=128,activation=tf.nn.tanh,dtype=tf.float32,name='lstm1')
    cell1= tf.nn.rnn_cell.DropoutWrapper(cell1,output_keep_prob=0.5)
    output1,states1 = tf.nn.dynamic_rnn(cell1,output0,dtype=tf.float32,swap_memory=True)


    cell2 = tf.nn.rnn_cell.LSTMCell(num_units=128,activation=tf.nn.tanh,dtype=tf.float32,name='lstm2')
    cell2= tf.nn.rnn_cell.DropoutWrapper(cell2,output_keep_prob=0.5)
    output2,states2 = tf.nn.dynamic_rnn(cell2,output1,dtype=tf.float32,swap_memory=True)

    attention_output, alphas = attention(output2, 50, return_alphas=True)
    drop = tf.nn.dropout(attention_output,keep_prob=0.5)

    attention_states = tf.multiply(drop,states2[0]) #CONTEXT INSTEAD OF HS

    cell3 = tf.layers.dense(units=128,activation=tf.nn.relu,inputs=attention_states,name='dense')
    cell4 = tf.layers.dense(units=6,activation=None,inputs=cell3,name='dense2')


    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=cell4,labels=y))
    optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
    training_op = optimizer.minimize(loss)
    
    correct_prediction = tf.equal(tf.argmax(cell4, 1), tf.argmax(y, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    #init = tf.global_variables_initializer()
    predict = cell4

    
############################################################################################################################    
    
    
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())

        for epoch_no in range(epochs):
            for index, offset in enumerate(range(0, X_train.shape[0], batch_size)):
                xs, ys = X_train[offset: offset + batch_size], y_train[offset: offset + batch_size]
                sess.run(training_op, feed_dict={X: xs, y: ys})
                y_pred=sess.run(predict,feed_dict={X:X_test})

                
            los, acc = sess.run([loss, accuracy], feed_dict = {X: xs, y: ys})
            val_los, val_acc = sess.run([loss, accuracy], feed_dict = {X: X_test, y: y_test})
            print (f'id {i} epoch {epoch_no} index {index}, Loss: {los:.4f}, accuracy = {acc:.4f}, Val_loss ={val_los:.4f} , val_acc ={val_acc:.4f}')
    current_confusion_Mat = result(y_pred,y_test)
    confusionMatrix.append(current_confusion_Mat)
    results.append((val_los,val_acc))
    tf.reset_default_graph()

        
```
